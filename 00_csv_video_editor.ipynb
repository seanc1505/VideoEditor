```python
data = pd.read_csv(csv_name)
data.fillna("", inplace = True)
# data = data[0:4]
data
```

```python
for index, row in data.iterrows():
    # print(index,row)
    temp_row = row
    # print(temp_row)
    for column in range(len(row)):
        # print(column)
        if row[column] =="":
            if index>0:
                temp_row[column] = prev_row[column]
            # print(temp_row)

        elif column>1 and isinstance(temp_row[column],str):
            time_list = temp_row[column].split(":",1)
            time = (float(time_list[0])*60) + float(time_list[1])
            temp_row[column] = time
        if column == 0:
            temp = str(row[column])
            if temp[-4:] != ".mp4":
                temp_row[column] = temp[:] + ".mp4"
            
    data.iloc[index] = temp_row
    # print(data.iloc[index])
    prev_row = row


data
```

```python
paddler_name_list = data['Paddler'].unique()
video_name_list = data['Video Name'].unique()
paddler_name_list
video_name_list
```

```python

for paddler in paddler_name_list:
    subclip_list = []
    export_video_name = ""
    video_length  = 0
    for video in video_name_list:
        
        # if not os.path.exists(source_video_path+video): 
        # print("Where is source video")
        
        video_clip = moviepy.VideoFileClip((source_video_path+"/"+video))
        for index, row in data.iterrows():
            if row['Paddler'] == paddler:
                if row['Video Name'] == video:
                    # Calculate full video length
                    current_subclip = video_clip.subclip(row["Start Time"], row["Stop Time"])
                    subclip_list.append(current_subclip)
    print(subclip_list)
    final_clip = concatenate_videoclips(subclip_list)
    export_video_name = paddler+"_"+export_video_name
    export_video_name =  filedialog.asksaveasfilename(filetypes = files, defaultextension = files,initialfile=export_video_name)
    
    # export_video_name = paddler + export_video_name
    # export_video_name = check_file_name(export_video_name,export_video_default_dir)
    final_clip.write_videofile(export_video_name)   
    # print(subclip_list)
    
```
